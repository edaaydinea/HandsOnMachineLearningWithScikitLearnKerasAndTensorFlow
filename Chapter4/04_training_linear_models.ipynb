## The Normal Equation

````python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
````

````python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
save_fig("generated_data_plot")
plt.show()
````

````python
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
````

This code is implementing the Normal Equation to compute the optimal parameters $\theta_{\text{best}}$ for a **linear regression** model. Let’s break it down:

### 1. `np.c_[np.ones((100, 1)), X]`:
- `np.c_[]`: This is used to concatenate arrays column-wise.
- `np.ones((100, 1))`: This creates a column vector of ones with 100 rows and 1 column. These ones represent the bias term $x_0 = 1$, added to the dataset $X$.
- `X`: The feature matrix with shape (100, n), where `n` is the number of features.
- The concatenation of `np.ones((100, 1))` with `X` results in a new matrix `X_b` of shape (100, n+1), where the first column is all ones (for the intercept term), and the rest are the original features.

### 2. `theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)`:
- `X_b.T`: This is the transpose of `X_b`, which flips the rows and columns.
- `X_b.T.dot(X_b)`: This computes the product of $X_b^T$ (transpose) and $X_b$, resulting in a square matrix.
- `np.linalg.inv(...)`: This computes the inverse of the matrix $X_b^T X_b$.
- `.dot(X_b.T).dot(y)`: This computes the dot product of the inverse matrix with $X_b^T$, followed by the dot product with $y$, the target vector (i.e., the output values for the linear regression).

### Explanation:
This code is solving for $\theta_{\text{best}}$ in the **Normal Equation** for linear regression:

$$
\theta_{\text{best}} = (X_b^T X_b)^{-1} X_b^T y
$$

Where:
- $X_b$ is the design matrix (with a bias term added).
- $y$ is the vector of target values.
- $\theta_{\text{best}}$ are the parameters that minimize the cost function in linear regression (least squares solution).

### Summary:
This code is computing the optimal values for the parameters $\theta$ (i.e., the weights) for a linear regression model using the Normal Equation method, which directly solves for the parameters without requiring gradient descent.

````python
theta_best
````

````python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
````

This code snippet is making **predictions** using a linear regression model that has already been trained. Let’s break it down step by step:

### 1. `X_new = np.array([[0], [2]])`:
- This defines a new input feature matrix `X_new` with two new instances, each containing one feature value:
  - The first instance has the feature value 0.
  - The second instance has the feature value 2.
- `X_new` has a shape of (2, 1), representing 2 data points with 1 feature each.

### 2. `X_new_b = np.c_[np.ones((2, 1)), X_new]`:
- This is the same step as before: adding a bias term (column of ones) to the new feature matrix `X_new`.
- `np.ones((2, 1))` creates a column vector of ones with 2 rows.
- `np.c_[np.ones((2, 1)), X_new]` concatenates the column of ones with `X_new`, resulting in `X_new_b`, which now has a shape of (2, 2). It looks like this:
  ```
  [[1. 0.]
   [1. 2.]]
  ```
  Here, the first column represents the bias term \(x_0 = 1\), and the second column represents the original feature values from `X_new`.

### 3. `y_predict = X_new_b.dot(theta_best)`:
- `theta_best` is the optimal parameter vector that was computed earlier using the Normal Equation.
- `X_new_b.dot(theta_best)` performs the dot product between `X_new_b` and `theta_best`, which results in predicted values (`y_predict`) for the new instances.
  
### 4. `y_predict` output:
- The result of the dot product gives the predicted values:
  ```
  array([[4.21509616],
         [9.75532293]])
  ```
- This means the predictions for the two new input values are:
  - For the instance where the feature value is 0, the predicted output is approximately 4.215.
  - For the instance where the feature value is 2, the predicted output is approximately 9.755.

### Summary:
- **Input**: Two new data points with feature values `0` and `2`.
- **Process**: The code adds a bias term and multiplies the resulting matrix by the model’s learned parameters (`theta_best`).
- **Output**: The predicted outputs (target values) are 4.215 and 9.755 for the inputs `0` and `2`, respectively.

The `LinearRegression` class is based on the `scipy.linalg.lstsq()` function (the name stands for "least squares"), which you could call directly:

````python
theta_best_svd, residuals, rank, s = np.linalg.lstsq(X_b, y, rcond=1e-6)
theta_best_svd
````

This code uses the **Singular Value Decomposition (SVD)** approach via `np.linalg.lstsq()` to solve for the optimal parameters $\theta_{\text{best}}$ in linear regression. It computes the least-squares solution for the system $X_b \theta = y$, where $X_b$ is the design matrix with the bias term added, and $y$ is the target vector. Let’s break down the code:

### 1. `np.linalg.lstsq(X_b, y, rcond=1e-6)`:
- **`np.linalg.lstsq`**: This function solves the linear least-squares problem, finding the parameters $\theta_{\text{best}}$ that minimize the squared differences between the observed target values $y$ and the predicted values from the model.
  
- **Parameters**:
  - `X_b`: The design matrix with the added bias term (shape $(m, n+1)$, where `m` is the number of samples and `n` is the number of features).
  - `y`: The target values, which the model is trying to predict (shape $(m, 1)$).
  - `rcond=1e-6`: This is a regularization parameter to deal with small singular values during the SVD computation. It helps avoid numerical instability.
  
- **Returns**:
  - **`theta_best_svd`**: The optimal solution vector, $\theta_{\text{best}}$, containing the intercept and coefficients.
  - **`residuals`**: The sum of squared residuals (differences between actual and predicted values). For overdetermined systems, it returns the residuals; for exact solutions, it’s an empty array.
  - **`rank`**: The effective rank of the matrix $X_b$.
  - **`s`**: The singular values of $X_b$.

### 2. `theta_best_svd`:
- This stores the computed optimal parameters (intercept and coefficients) for the linear regression model. The result:
  ```
  array([[4.21509616],
         [2.77011339]])
  ```
  indicates that:
  - **Intercept**: $4.215$
  - **Coefficient**: $2.770$ (the slope of the feature).

This means the model is:
$$
y = 4.215 + 2.770 \cdot x
$$

### Summary:
- `np.linalg.lstsq` solves the least-squares problem using the **SVD** approach to find the optimal linear regression parameters.
- The result is the same as in previous methods: an intercept of `4.215` and a coefficient of `2.770`, which are the parameters for the linear regression model.

## Stochastic Gradient Descent

````python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
````

````python
n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # random initialization

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    # not shown in the book
            y_predict = X_new_b.dot(theta)           # not shown
            style = "b-" if i > 0 else "r--"         # not shown
            plt.plot(X_new, y_predict, style)        # not shown
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 # not shown

plt.plot(X, y, "b.")                                 # not shown
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, 0, 15])                              # not shown
save_fig("sgd_plot")                                 # not shown
plt.show()                                           # not shown
````

This code implements **Stochastic Gradient Descent (SGD)** to optimize the parameters $\theta$ for a linear regression model. SGD is an iterative optimization algorithm where the model parameters are updated in small steps based on individual training examples rather than the entire dataset (which would be the case with standard Gradient Descent). Here’s a breakdown of the key components:

### 1. **Hyperparameters**:
   - `n_epochs = 50`: The number of complete passes (epochs) through the training data.
   - `t0, t1 = 5, 50`: These are hyperparameters used to control the learning rate schedule.

### 2. **Learning Schedule**:
   - `learning_schedule(t)` defines how the learning rate ($\eta$) changes over time.
   - The learning rate is updated with:
     $$
     \eta(t) = \frac{t_0}{t + t_1}
     $$
     As the algorithm runs and `t` increases, the learning rate decreases over time. This allows the model to take larger steps initially (to converge faster) and smaller steps later (to fine-tune the model).

### 3. **Random Initialization**:
   - `theta = np.random.randn(2,1)` randomly initializes the model parameters ($\theta$) as a 2x1 vector (for one feature plus the bias term). This step provides a starting point for the optimization.

### 4. **Main Loop**:
   - `for epoch in range(n_epochs)`: The outer loop runs for 50 epochs (passes over the entire training set).
   - `for i in range(m)`: The inner loop runs for each training instance in the dataset. This loop implements **stochastic** updates by selecting random samples.

### 5. **Random Sampling**:
   - `random_index = np.random.randint(m)`: A random index is chosen to select one training example at a time. This randomness is key to Stochastic Gradient Descent (as opposed to regular gradient descent, which processes all samples simultaneously).
   - `xi = X_b[random_index:random_index+1]`: Extracts the randomly chosen training example from the design matrix $X_b$. The slicing keeps the shape of the array intact as a 2D array.
   - `yi = y[random_index:random_index+1]`: Extracts the corresponding target value.

### 6. **Gradient Calculation**:
   - `gradients = 2 * xi.T.dot(xi.dot(theta) - yi)`: Computes the gradient of the cost function with respect to $\theta$ for the randomly chosen training example.
     - `xi.dot(theta) - yi`: Computes the error (difference between predicted value and actual value).
     - `xi.T.dot(...)`: Multiplies the error by the features of the randomly selected instance (this gives the gradient).
     - The factor of `2` comes from the derivative of the squared loss function used in linear regression.

### 7. **Learning Rate Update**:
   - `eta = learning_schedule(epoch * m + i)`: The learning rate is computed for the current iteration using the `learning_schedule` function. The learning rate decreases as the number of iterations increases.

### 8. **Parameter Update**:
   - `theta = theta - eta * gradients`: The model parameters $\theta$ are updated by moving in the opposite direction of the gradient (which points in the direction of increasing error). The step size is controlled by the learning rate $\eta$.

### Summary:
- The code uses **Stochastic Gradient Descent** to update the model parameters $\theta$ for a linear regression model.
- At each iteration, a single random training example is used to compute the gradient and update $\theta$.
- The learning rate decreases over time according to the `learning_schedule`, allowing the model to converge more smoothly.

## Mini-batch gradient descent

````python
theta_path_mgd = []

n_iterations = 50 #the number of complete passes through the training set, also called epochs
minibatch_size = 20 # these are hyperparameters used to control the learning rate schedule

np.random.seed(42)
theta = np.random.randn(2,1)  # randomly initializes the model parameters as a 2x1 vector (for one feature plus the bias term). This step provides a starting point for the optimization.

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
# The outer loop runs for 50 epochs(passes over the entire training set)
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    # The inner loop runs for 100 iterations, each iteration processes 20 instances (minibatch_size)
    for i in range(0, m, minibatch_size):
        t += 1
        # Extracts the randomly chose training example from the design matrix X_b. The slicing keeps the shape of the array infact as a 2D array.
        xi = X_b_shuffled[i:i+minibatch_size]
        # Extracts the correscponding target value.
        yi = y_shuffled[i:i+minibatch_size]
        # Computes the gradient of the cost function with respect to theta for the randomly chose training example.
        # xi.dot(theta) - yi : computes the error (difference between predicted value and actual value)
        # xi.T.dot(...) : Multiplies the error by the features of the randomly selected instance (this given the gradient)
        # 2/minibatch_size : scales the gradient by the number of instances in the minibatch
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        # Computes the learning rate for this iteration
        eta = learning_schedule(t)
        # Updates the model parameters
        theta = theta - eta * gradients
        # Appends the updated model parameters to the theta_path_mgd list
        theta_path_mgd.append(theta)
````

````python
theta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)
````

````python
plt.figure(figsize=(7,4))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
save_fig("gradient_descent_paths_plot")
plt.show()
````

````python
# The PolynomialFeatures class transforms an array containing n features into an array containing (n+d)!/(d!n!) features, where d is the degree of the polynomial.
from sklearn.preprocessing import PolynomialFeatures

# The following code creates a second-degree polynomial using the PolynomialFeatures class, then fits a LinearRegression model to the transformed training data.
poly_features = PolynomialFeatures(degree=2, include_bias=False)

# The fit_transform() method transforms the input data, X, into a new array containing the original features and the square of each feature.
X_poly = poly_features.fit_transform(X)

X[0]
````

````python
X_poly[0]
````

````python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train) + 1):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
````

````python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m, 1)

X_train, X_val, y_train, y_val = train_test_split(X[:50], y[:50].ravel(), test_size=0.5, random_state=10)
````

````python
from copy import deepcopy

poly_scaler = Pipeline([
        ("poly_features", PolynomialFeatures(degree=90, include_bias=False)),
        ("std_scaler", StandardScaler())
    ])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)

sgd_reg = SGDRegressor(max_iter=1, tol=None, warm_start=True,
                       penalty=None, learning_rate="constant", eta0=0.0005, random_state=42)

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)  # continues where it left off
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val, y_val_predict)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = deepcopy(sgd_reg)
````

````python
t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5

t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)
T = np.c_[t1.ravel(), t2.ravel()]
Xr = np.array([[1, 1], [1, -1], [1, 0.5]])
yr = 2 * Xr[:, :1] + 0.5 * Xr[:, 1:]

J = (1/len(Xr) * np.sum((T.dot(Xr.T) - yr.T)**2, axis=1)).reshape(t1.shape)

N1 = np.linalg.norm(T, ord=1, axis=1).reshape(t1.shape)
N2 = np.linalg.norm(T, ord=2, axis=1).reshape(t1.shape)

t_min_idx = np.unravel_index(np.argmin(J), J.shape)
t1_min, t2_min = t1[t_min_idx], t2[t_min_idx]

t_init = np.array([[0.25], [-1]])
````

````python
def bgd_path(theta, X, y, l1, l2, core = 1, eta = 0.05, n_iterations = 200):
    path = [theta]
    for iteration in range(n_iterations):
        gradients = core * 2/len(X) * X.T.dot(X.dot(theta) - y) + l1 * np.sign(theta) + l2 * theta
        theta = theta - eta * gradients
        path.append(theta)
    return np.array(path)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10.1, 8))
for i, N, l1, l2, title in ((0, N1, 2., 0, "Lasso"), (1, N2, 0,  2., "Ridge")):
    JR = J + l1 * N1 + l2 * 0.5 * N2**2
    
    tr_min_idx = np.unravel_index(np.argmin(JR), JR.shape)
    t1r_min, t2r_min = t1[tr_min_idx], t2[tr_min_idx]

    levelsJ=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(J) - np.min(J)) + np.min(J)
    levelsJR=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(JR) - np.min(JR)) + np.min(JR)
    levelsN=np.linspace(0, np.max(N), 10)
    
    path_J = bgd_path(t_init, Xr, yr, l1=0, l2=0)
    path_JR = bgd_path(t_init, Xr, yr, l1, l2)
    path_N = bgd_path(np.array([[2.0], [0.5]]), Xr, yr, np.sign(l1)/3, np.sign(l2), core=0)

    ax = axes[i, 0]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, N / 2., levels=levelsN)
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.set_title(r"$\ell_{}$ penalty".format(i + 1), fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)
    ax.set_ylabel(r"$\theta_2$", fontsize=16, rotation=0)

    ax = axes[i, 1]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, JR, levels=levelsJR, alpha=0.9)
    ax.plot(path_JR[:, 0], path_JR[:, 1], "w-o")
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.plot(t1r_min, t2r_min, "rs")
    ax.set_title(title, fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)

save_fig("lasso_vs_ridge_plot")
plt.show()
````

## 12. Batch Gradient Descent with early stopping for Softmax Regression
(without using Scikit-Learn)

Let's start by loading the data. We will just reuse the Iris dataset we loaded earlier.

````python
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]
````

We need to add the bias term for every instance ($x_0 = 1$):

````python
X_with_bias = np.c_[np.ones([len(X), 1]), X]
````

And let's set the random seed so the output of this exercise solution is reproducible:

````python
np.random.seed(2042)
````

The easiest option to split the dataset into a training set, a validation set and a test set would be to use Scikit-Learn's `train_test_split()` function, but the point of this exercise is to try understand the algorithms by implementing them manually. So here is one possible implementation:

````python
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_with_bias)

test_size = int(total_size * test_ratio)
validation_size = int(total_size * validation_ratio)
train_size = total_size - test_size - validation_size

rnd_indices = np.random.permutation(total_size)

X_train = X_with_bias[rnd_indices[:train_size]]
y_train = y[rnd_indices[:train_size]]
X_valid = X_with_bias[rnd_indices[train_size:-test_size]]
y_valid = y[rnd_indices[train_size:-test_size]]
X_test = X_with_bias[rnd_indices[-test_size:]]
y_test = y[rnd_indices[-test_size:]]
````

The targets are currently class indices (0, 1 or 2), but we need target class probabilities to train the Softmax Regression model. Each instance will have target class probabilities equal to 0.0 for all classes except for the target class which will have a probability of 1.0 (in other words, the vector of class probabilities for ay given instance is a one-hot vector). Let's write a small function to convert the vector of class indices into a matrix containing a one-hot vector for each instance:

````python
def to_one_hot(y):
    n_classes = y.max() + 1
    m = len(y)
    Y_one_hot = np.zeros((m, n_classes))
    Y_one_hot[np.arange(m), y] = 1
    return Y_one_hot
````

Let's test this function on the first 10 instances:

````python
y_train[:10]
````

````python
to_one_hot(y_train[:10])
````

Looks good, so let's create the target class probabilities matrix for the training set and the test set:

````python
Y_train_one_hot = to_one_hot(y_train)
Y_valid_one_hot = to_one_hot(y_valid)
Y_test_one_hot = to_one_hot(y_test)
````

Now let's implement the Softmax function. Recall that it is defined by the following equation:

$\sigma\left(\mathbf{s}(\mathbf{x})\right)_k = \dfrac{\exp\left(s_k(\mathbf{x})\right)}{\sum\limits_{j=1}^{K}{\exp\left(s_j(\mathbf{x})\right)}}$

````python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = np.sum(exps, axis=1, keepdims=True)
    return exps / exp_sums
````

We are almost ready to start training. Let's define the number of inputs and outputs:

````python
n_inputs = X_train.shape[1] # == 3 (2 features plus the bias term)
n_outputs = len(np.unique(y_train))   # == 3 (3 iris classes)
````

Now here comes the hardest part: training! Theoretically, it's simple: it's just a matter of translating the math equations into Python code. But in practice, it can be quite tricky: in particular, it's easy to mix up the order of the terms, or the indices. You can even end up with code that looks like it's working but is actually not computing exactly the right thing. When unsure, you should write down the shape of each term in the equation and make sure the corresponding terms in your code match closely. It can also help to evaluate each term independently and print them out. The good news it that you won't have to do this everyday, since all this is well implemented by Scikit-Learn, but it will help you understand what's going on under the hood.

So the equations we will need are the cost function:

$J(\mathbf{\Theta}) =
- \dfrac{1}{m}\sum\limits_{i=1}^{m}\sum\limits_{k=1}^{K}{y_k^{(i)}\log\left(\hat{p}_k^{(i)}\right)}$

And the equation for the gradients:

$\nabla_{\mathbf{\theta}^{(k)}} \, J(\mathbf{\Theta}) = \dfrac{1}{m} \sum\limits_{i=1}^{m}{ \left ( \hat{p}^{(i)}_k - y_k^{(i)} \right ) \mathbf{x}^{(i)}}$

Note that $\log\left(\hat{p}_k^{(i)}\right)$ may not be computable if $\hat{p}_k^{(i)} = 0$. So we will add a tiny value $\epsilon$ to $\log\left(\hat{p}_k^{(i)}\right)$ to avoid getting `nan` values.

````python
eta = 0.01
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    if iteration % 500 == 0:
        loss = -np.mean(np.sum(Y_train_one_hot * np.log(Y_proba + epsilon), axis=1))
        print(iteration, loss)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error)
    Theta = Theta - eta * gradients
````

And that's it! The Softmax model is trained. Let's look at the model parameters:

````python
Theta
````

Let's make predictions for the validation set and check the accuracy score:

````python
logits = X_valid.dot(Theta)
Y_proba = softmax(logits)
y_predict = np.argmax(Y_proba, axis=1)

accuracy_score = np.mean(y_predict == y_valid)
accuracy_score
````

Well, this model looks pretty good. For the sake of the exercise, let's add a bit of $\ell_2$ regularization. The following training code is similar to the one above, but the loss now has an additional $\ell_2$ penalty, and the gradients have the proper additional term (note that we don't regularize the first element of `Theta` since this corresponds to the bias term). Also, let's try increasing the learning rate `eta`.

````python
eta = 0.1
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    if iteration % 500 == 0:
        xentropy_loss = -np.mean(np.sum(Y_train_one_hot * np.log(Y_proba + epsilon), axis=1))
        l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
        loss = xentropy_loss + alpha * l2_loss
        print(iteration, loss)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients
````

Because of the additional $\ell_2$ penalty, the loss seems greater than earlier, but perhaps this model will perform better? Let's find out:

````python
logits = X_valid.dot(Theta)
Y_proba = softmax(logits)
y_predict = np.argmax(Y_proba, axis=1)

accuracy_score = np.mean(y_predict == y_valid)
accuracy_score
````

Cool, perfect accuracy! We probably just got lucky with this validation set, but still, it's pleasant.

Now let's add early stopping. For this we just need to measure the loss on the validation set at every iteration and stop when the error starts growing.

````python
eta = 0.1 
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter
best_loss = np.infty

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients

    logits = X_valid.dot(Theta)
    Y_proba = softmax(logits)
    xentropy_loss = -np.mean(np.sum(Y_valid_one_hot * np.log(Y_proba + epsilon), axis=1))
    l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
    loss = xentropy_loss + alpha * l2_loss
    if iteration % 500 == 0:
        print(iteration, loss)
    if loss < best_loss:
        best_loss = loss
    else:
        print(iteration - 1, best_loss)
        print(iteration, loss, "early stopping!")
        break
````

````python
logits = X_valid.dot(Theta)
Y_proba = softmax(logits)
y_predict = np.argmax(Y_proba, axis=1)

accuracy_score = np.mean(y_predict == y_valid)
accuracy_score
````

Still perfect, but faster.
